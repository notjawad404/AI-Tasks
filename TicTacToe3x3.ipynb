```python
import random

allowednumber = [0,1,2,3,4,5,6,7,8]

game = [[0,1,2], [3,4,5], [6,7,8]]

rows = 3
col = 3
print("Tic Tac Toe 3x3 Player vs Jarvis")
def printGame():
  for i in range(rows):
    print("\n|---|---|---|")
    print("|", end="")
    for j in range(col):
      print("", game[i][j], end=" |")
  print("\n|---|---|---|")


printGame()
print("X")
print("Y")

def newArray(num, turn):
  
  if(num== 0):
    game[0][0] = turn
  if(num== 1):
    game[0][1] = turn
  if(num== 2):
    game[0][2] = turn
  if(num== 3):
    game[1][0] = turn
  if(num== 4):
    game[1][1] = turn
  if(num== 5):
    game[1][2] = turn
  if(num== 6):
    game[2][0] = turn
  if(num== 7):
    game[2][1] = turn
  if(num== 8):
    game[2][2] = turn
  
def checkWinner(game):
  if(game[0][0] == 'X' and game[0][1] == 'X' and game[0][2] == 'X'):
    print("Jarvis has won!")
    return "X"
  elif(game[0][0] == 'Y' and game[0][1] == 'Y' and game[0][2] == 'Y'):
    print("You has won!")
    return "Y"
  elif(game[1][0] == 'X' and game[1][1] == 'X' and game[1][2] == 'X'):
    print("Jarvis has won!")
    return "X"
  elif(game[1][0] == 'Y' and game[1][1] == 'Y' and game[1][2] == 'Y'):
    print("You has won!")
    return "Y"
  elif(game[2][0] == 'X' and game[2][1] == 'X' and game[2][2] == 'X'):
    print("Jarvis has won!")
    return "X"
  elif(game[2][0] == 'Y' and game[2][1] == 'Y' and game[2][2] == 'Y'):
    print("You has won!")
    return "Y"
  # Columns 
  if(game[0][0] == 'X' and game[1][0] == 'X' and game[2][0] == 'X'):
    print("Jarvis has won!")
    return "X"
  elif(game[0][0] == 'Y' and game[1][0] == 'Y' and game[2][0] == 'Y'):
    print("You has won!")
    return "Y"
  elif(game[0][1] == 'X' and game[1][1] == 'X' and game[2][1] == 'X'):
    print("Jarvis has won!")
    return "X"
  elif(game[0][1] == 'Y' and game[1][1] == 'Y' and game[2][1] == 'Y'):
    print("You has won!")
    return "Y"
  elif(game[0][2] == 'X' and game[1][2] == 'X' and game[2][2] == 'X'):
    print("Jarvis has won!")
    return "X"
  elif(game[0][2] == 'Y' and game[1][2] == 'Y' and game[2][2] == 'Y'):
    print("You has won!")
    return "Y"
  #Diagonal
  elif(game[0][0] == 'X' and game[1][1] == 'X' and game[2][2] == 'X'):
    print("Jarvis has won!")
    return "X"
  elif(game[0][0] == 'Y' and game[1][1] == 'Y' and game[2][2] == 'Y'):
    print("You has won!")  
    return "Y"
  elif(game[0][2] == 'X' and game[1][1] == 'X' and game[2][0] == 'X'):
    print("Jarvis has won!")  
    return "X"
  elif(game[0][2] == 'Y' and game[1][1] == 'Y' and game[2][0] == 'Y'):
    print("You has won!") 
    return "Y" 
  else:
    return "DRAW"
gameLoop = True
turnCounter = 0

while(gameLoop == True):
  if(turnCounter %2 == 1):
    printGame()
    number = int(input("Choose a number between 0 - 8: "))
    if(number >=  0 or number <=8 ):
      if( number in allowednumber):
        newArray(number, "Y")
        allowednumber.remove(number)
        turnCounter += 1
      else:
        print("Invalid Number, Try again")
    else:
      print("Invalid Number, Try again")

  else:
    while(True):
         JarvisChoice = random.choice(allowednumber)
         print("\nJarvis choosed: ", JarvisChoice)
         if(JarvisChoice in allowednumber):
           newArray(JarvisChoice, "X")
           allowednumber.remove(JarvisChoice)
           turnCounter +=1
           break
    winner = checkWinner(game)
    if(winner != "DRAW"):
      print("\nGame over!")
      break
```
